```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.optimize import curve_fit
```

```python
file = np.genfromtxt('./drcPhot10Nov/catDir_HOROLOGIUM-I/HOROLOGIUM-I_fullCat.dat',names=True)
```

```python
def func(x, a, b, c):
    return a * np.exp(b * x) + c
```

```python
x = file['magr_f606w']
y = file['stdev_f606w']
```

```python
popt, pcov = curve_fit(func, x, y)
```

```python
popt0, _ = curve_fit(func, x, y,p0=popt)
```
