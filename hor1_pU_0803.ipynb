```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = np.genfromtxt('photUtils0820/HOROLOGIUM-I_filtMatch_pU.dat',\
                     names=True)
```

```python
fig, ax = plt.subplots(figsize=(4,6))

ax.scatter(file['magr_f606w']-file['magr_f814w'],file['magr_f606w'],s=5)

ax.set_ylim(28,17)
ax.set_xlim(-2,1.5)
ax.set_xlabel('F606W-F814W')
ax.set_ylabel('F606W')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.scatter(file['xcenter_f606w'],file['ycenter_f606w'],s=5)


plt.show()
```

```python
file2 = np.genfromtxt('photUtils0820/HOROLOGIUM-I_aperPuMatch_pU.dat',\
                      names=True)
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.scatter(file['xcenter_f606w'],file['ycenter_f606w'],s=5)
ax.scatter(file2['xcenter_f606w'],file2['ycenter_f606w'],s=5)


plt.show()
```
