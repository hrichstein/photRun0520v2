```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sepCat = np.genfromtxt('catsSept21/toSendSept21/catsSept21_HOROLOGIUM-I.dat',
                      names=True)
novCat = np.genfromtxt('forElena16Nov/HOROLOGIUM-I_cat16Nov.dat',names=True)
julCat = np.genfromtxt('catDir0720/HOROLOGIUM-I_sourceList0720.dat',names=True)

```

```python
sepCat2 = np.genfromtxt('catsSept21/toSendSept21/catsSept21_PEGASUS-III.dat',
                      names=True)
novCat2 = np.genfromtxt('forElena16Nov/PEGASUS-III_cat16Nov.dat',names=True)
julCat2 = np.genfromtxt('catDir0720/PEGASUS-III_sourceList0720.dat',names=True)
```

```python
cutIdx = np.logical_and(junCat['class_Star_i']>=0.95,junCat['class_Star_v']>=0.95)
junCatcut = junCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(novCat['star_f606w']==1,novCat['star_f814w']==1),
                       np.logical_and(novCat['err_f606w']<0.05,novCat['err_f814w']<0.05))
novCatcut = novCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(julCat['c_star_f606w']>=0.8,julCat['c_star_f814w']>=0.95),
                       np.logical_and(julCat['magZPTerr_f606w']<0.05,julCat['magZPTerr_f814w']<0.05))
julCatcut = julCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(julCat2['c_star_f606w']>=0.8,julCat2['c_star_f814w']>=0.8),
                       np.logical_and(julCat2['magZPTerr_f606w']<0.1,julCat2['magZPTerr_f814w']<0.1))
julCatcut2 = julCat2[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(sepCat['c_star_f606w']>=0.95,sepCat['c_star_f814w']>=0.95),
                       np.logical_and(sepCat['magZPTerr_f606w']<0.05,sepCat['magZPTerr_f814w']<0.05))
sepCatcut = sepCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(sepCat2['c_star_f606w']>=0.95,sepCat2['c_star_f814w']>=0.95),
                       np.logical_and(sepCat2['magZPTerr_f606w']<0.05,sepCat2['magZPTerr_f814w']<0.05))
sepCatcut2 = sepCat2[cutIdx]
```

```python
cutIdx2 = np.logical_and(np.logical_and(novCat2['star_f606w']==1,novCat2['star_f814w']==1),
                       np.logical_and(novCat2['err_f606w']<0.05,novCat2['err_f814w']<0.05))
novCatcut2 = novCat2[cutIdx2]
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(16,6),sharex=True,sharey=True)

ax1.scatter(junCat['magAbs_v']-junCat['magAbs_i'],junCat['magAbs_v'],s=5)
ax1.scatter(junCatcut['magAbs_v']-junCatcut['magAbs_i'],junCatcut['magAbs_v'],s=5)
ax2.scatter(julCat['magZPT_f606w']-julCat['magZPT_f814w'],julCat['magZPT_f606w']-19.6,s=5)
ax2.scatter(julCatcut['magZPT_f606w']-julCatcut['magZPT_f814w'],julCatcut['magZPT_f606w']-19.6,s=5)
ax3.scatter(sepCat['magZPT_f606w']-sepCat['magZPT_f814w'],sepCat['magZPT_f606w']-19.6,s=5)
ax3.scatter(sepCatcut['magZPT_f606w']-sepCatcut['magZPT_f814w'],sepCatcut['magZPT_f606w']-19.6,s=5)
ax4.scatter(novCat['magr_f606w']-novCat['magr_f814w'],novCat['magr_f606w']-19.6,s=5)
ax4.scatter(novCatcut['magr_f606w']-novCatcut['magr_f814w'],novCatcut['magr_f606w']-19.6,s=5)

ax1.set_title('June')
ax2.set_title('July')
ax3.set_title('September')
ax4.set_title('November')
ax1.set_ylim(8,-2)
ax1.set_xlim(-1.,0.25)

plt.subplots_adjust(wspace=0)
plt.show()
```
