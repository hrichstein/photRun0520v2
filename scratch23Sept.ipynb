```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hydra2 = np.genfromtxt('catsSept21/toSendSept21/catsSept21_HYDRA-II.dat',names=True)
```

```python
tuc2 = np.genfromtxt('catsSept21/toSendSept21/TUCANA-II_all_source.dat',names=True)
```

```python
fig,ax = plt.subplots()


ax.scatter(hydra2['magZPT_f606w'],hydra2['stdev_f606w'])
ax.scatter(hydra2['magZPT_f606w'],hydra2['magZPTerr_f606w'])


plt.show()
```

```python
fig,ax = plt.subplots()


ax.scatter(tuc2['magZPT_f606w'],tuc2['stdev_f606w'])
ax.scatter(tuc2['magZPT_f606w'],tuc2['magZPTerr_f606w'])

ax.scatter(tuc2['magZPT_f606w'][:50],tuc2['magZPTerr_f606w'][:50])
ax.scatter(tuc2['magZPT_f606w'][-50:],tuc2['magZPTerr_f606w'][-50:])


plt.show()
```

```python
fig,ax = plt.subplots()

ax.scatter(hydra2['magZPT_f606w'],hydra2['magZPT_f606w']-hydra2['mean_f606w'])
ax.scatter(tuc2['magZPT_f606w'],tuc2['magZPT_f606w']-tuc2['mean_f606w'])
ax.scatter(tuc2['magZPT_f606w'][:50],tuc2['magZPT_f606w'][:50]-tuc2['mean_f606w'][:50])

plt.show()
```
