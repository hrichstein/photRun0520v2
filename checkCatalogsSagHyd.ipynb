```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sepCat = np.genfromtxt('catsSept21/toSendSept21/catsSept21_SAGITTARIUS-II.dat',
                      names=True)
novCat = np.genfromtxt('forElena16Nov/SAGITTARIUS-II_cat16Nov.dat',names=True)
julCat = np.genfromtxt('catDir0720/SAGITTARIUS-II_sourceList0720.dat',names=True)

```

```python
sepCat2 = np.genfromtxt('catsSept21/toSendSept21/catsSept21_HYDRA-II.dat',
                      names=True)
novCat2 = np.genfromtxt('forElena16Nov/HYDRA-II_cat16Nov.dat',names=True)
julCat2 = np.genfromtxt('catDir0720/HYDRA-II_sourceList0720.dat',names=True)
```

```python
cutIdx = np.logical_and(np.logical_and(novCat['star_f606w']==1,novCat['star_f814w']==1),
                       np.logical_and(novCat['err_f606w']<0.05,novCat['err_f814w']<0.05))
novCatcut = novCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(julCat['c_star_f606w']>=0.95,julCat['c_star_f814w']>=0.95),
                       np.logical_and(julCat['magZPTerr_f606w']<0.05,julCat['magZPTerr_f814w']<0.05))
julCatcut = julCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(julCat2['c_star_f606w']>=0.95,julCat2['c_star_f814w']>=0.95),
                       np.logical_and(julCat2['stdev_f606w']<0.05,julCat2['stdev_f814w']<0.05))
julCatcut2 = julCat2[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(sepCat['c_star_f606w']>=0.95,sepCat['c_star_f814w']>=0.95),
                       np.logical_and(sepCat['magZPTerr_f606w']<0.05,sepCat['magZPTerr_f814w']<0.05))
sepCatcut = sepCat[cutIdx]
```

```python
cutIdx = np.logical_and(np.logical_and(sepCat2['c_star_f606w']>=0.95,sepCat2['c_star_f814w']>=0.95),
                       np.logical_and(sepCat2['stdev_f606w']<0.05,sepCat2['stdev_f814w']<0.05))
sepCatcut2 = sepCat2[cutIdx]
```

```python
cutIdx2 = np.logical_and(np.logical_and(novCat2['star_f606w']==1,novCat2['star_f814w']==1),
                       np.logical_and(novCat2['err_f606w']<0.05,novCat2['err_f814w']<0.05))
novCatcut2 = novCat2[cutIdx2]
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(13,6),sharex=True,sharey=True)

ax2.scatter(sepCat['magZPT_f606w']-sepCat['magZPT_f814w'],sepCat['magZPT_f606w']-19.1,s=5)
ax2.scatter(sepCatcut['magZPT_f606w']-sepCatcut['magZPT_f814w'],sepCatcut['magZPT_f606w']-19.1,s=5)
ax3.scatter(novCat['magr_f606w']-novCat['magr_f814w'],novCat['magr_f606w']-19.1,s=5)
ax1.scatter(julCat['magZPT_f606w']-julCat['magZPT_f814w'],julCat['magZPT_f606w']-19.1,s=5)
ax1.scatter(julCatcut['magZPT_f606w']-julCatcut['magZPT_f814w'],julCatcut['magZPT_f606w']-19.1,s=5)
ax3.scatter(novCatcut['magr_f606w']-novCatcut['magr_f814w'],novCatcut['magr_f606w']-19.1,s=5)

ax1.set_title('July')
ax2.set_title('September')
ax3.set_title('November')
ax1.set_ylim(9,-2)
ax1.set_xlim(-1.,0.25)
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(9,6),sharex=True,sharey=True)

# ax2.scatter(sepCat['magZPT_f606w']-sepCat['magZPT_f814w'],sepCat['magZPT_f606w']-19.1,s=5)
# ax2.scatter(sepCatcut['magZPT_f606w']-sepCatcut['magZPT_f814w'],sepCatcut['magZPT_f606w']-19.1,s=5)
ax2.scatter(novCat['magr_f606w']-novCat['magr_f814w'],novCat['magr_f606w']-19.1,s=5,label='Full')
ax1.scatter(julCat['magZPT_f606w']-julCat['magZPT_f814w'],julCat['magZPT_f606w']-19.1,s=5,label='Full')
ax1.scatter(julCatcut['magZPT_f606w']-julCatcut['magZPT_f814w'],julCatcut['magZPT_f606w']-19.1,s=5,label='Cut')
ax2.scatter(novCatcut['magr_f606w']-novCatcut['magr_f814w'],novCatcut['magr_f606w']-19.1,s=5,label='Cut')

ax1.set_title('July')
# ax2.set_title('September')
ax2.set_title('November')
ax1.set_ylim(9,-1)
ax1.set_xlim(-1.25,0.45)

ax1.set_xlabel('F606W-F814W')
ax2.set_xlabel('F606W-F814W')
ax1.set_ylabel('F606W')

ax1.legend()
ax2.legend()

plt.subplots_adjust(wspace=0)
plt.savefig('sagII_comp.png',bbox_inches='tight',dpi=660)
plt.show()
```
