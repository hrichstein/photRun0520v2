```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
dat = np.genfromtxt('catRawMags1305/catDir/HOROLOGIUM-I_F606W_at_long.dat',names=True)
dat2 = np.genfromtxt('catRawMags1305/catDir/HOROLOGIUM-I_F814W_at_long.dat',names=True)
```

```python
mag_std = np.nanstd( np.array([dat['mag1'],dat['mag2'],dat['mag3'],dat['mag4']]),axis=0)
mag_std2 = np.nanstd( np.array([dat2['mag1'],dat2['mag2'],dat2['mag3'],dat2['mag4']]),axis=0)
```

```python
median_mag = np.nanmedian( np.array([dat['mag1'],dat['mag2'],dat['mag3'],dat['mag4']]),axis=0)
median_mag2 = np.nanmedian( np.array([dat2['mag1'],dat2['mag2'],dat2['mag3'],dat2['mag4']]),axis=0)
```

```python
def dist(x0,y0,x1,y1):
    xdiff = x0-x1
    ydiff = y0-y1
    
    dist = np.sqrt(xdiff**2 + ydiff**2)
    
    return dist

def dist_std(dist1,dist2,dist3):

    return np.nanstd(np.array([dist1,dist2,dist3]),axis=0)
```

```python
dist1 = dist(dat['xt1'],dat['yt1'],dat['xt2'],dat['yt2'])
dist2 = dist(dat['xt1'],dat['yt1'],dat['xt3'],dat['yt3'])
dist3 = dist(dat['xt1'],dat['yt1'],dat['xt4'],dat['yt4'])

pix_std = dist_std(dist1,dist2,dist3)
```

```python
dist1 = dist(dat2['xt1'],dat2['yt1'],dat2['xt2'],dat2['yt2'])
dist2 = dist(dat2['xt1'],dat2['yt1'],dat2['xt3'],dat2['yt3'])
dist3 = dist(dat2['xt1'],dat2['yt1'],dat2['xt4'],dat2['yt4'])

pix_std2 = dist_std(dist1,dist2,dist3)
```

```python
dist1 = dist(dat['ra1'],dat['dec1'],dat['ra2'],dat['dec2'])
dist2 = dist(dat['ra1'],dat['dec1'],dat['ra3'],dat['dec3'])
dist3 = dist(dat['ra1'],dat['dec1'],dat['ra4'],dat['dec4'])

coord_std = dist_std(dist1,dist2,dist3)
```

```python
dist1 = dist(dat2['ra1'],dat2['dec1'],dat2['ra2'],dat2['dec2'])
dist2 = dist(dat2['ra1'],dat2['dec1'],dat2['ra3'],dat2['dec3'])
dist3 = dist(dat2['ra1'],dat2['dec1'],dat2['ra4'],dat2['dec4'])

coord_std2 = dist_std(dist1,dist2,dist3)
```

```python
dist1 = dist(dat['wcsRA'],dat['wcsDEC'],dat['wra2'],dat['wdec2'])
dist2 = dist(dat['wcsRA'],dat['wcsDEC'],dat['wra3'],dat['wdec3'])
dist3 = dist(dat['wcsRA'],dat['wcsDEC'],dat['wra4'],dat['wdec4'])

wCoord_std = dist_std(dist1,dist2,dist3)
```

```python
dist1 = dist(dat2['wcsRA'],dat2['wcsDEC'],dat2['wra2'],dat2['wdec2'])
dist2 = dist(dat2['wcsRA'],dat2['wcsDEC'],dat2['wra3'],dat2['wdec3'])
dist3 = dist(dat2['wcsRA'],dat2['wcsDEC'],dat2['wra4'],dat2['wdec4'])

wCoord_std2 = dist_std(dist1,dist2,dist3)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,5))

ax1.scatter(mag_std,pix_std)
ax2.scatter(mag_std2,pix_std2)

ax1.set_xlabel('F606W Mag Err')
ax1.set_ylabel('F606W Pix Pos Err')

ax2.set_xlabel('F814W Mag Err')
ax2.set_ylabel('F814W Pix Pos Err')

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,5),sharey=True)

ax1.scatter(mag_std,coord_std)
ax2.scatter(mag_std,wCoord_std)

ax1.set_xlabel('F606W Mag Err')
ax1.set_ylabel('F606W SE Coord Err')

ax2.set_xlabel('F606W Mag Err')
ax2.set_ylabel('F606W WCS Coord Err')

ax1.set_ylim(-1e-4,2e-4)

plt.show()
```
