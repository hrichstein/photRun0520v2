```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
file = 'drcPhot02Nov/catDir_HOROLOGIUM-I/HOROLOGIUM-I_matchedDRCfilt.dat'
```

```python
data = np.genfromtxt(file,names=True)
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,7),sharey=True)

im1 = ax1.scatter(data['magr_f606w']-data['magr_f814w'],data['magr_f606w'],
           c=data['sharpness_f606w'],s=10)
im2=ax2.scatter(data['magr_f606w']-data['magr_f814w'],data['magr_f606w'],
           c=data['roundness1_f606w'],s=10)
ax1.set_ylim(27,18)
ax1.set_xlim(-2,2)
ax2.set_xlim(-2,2)


divider1 = make_axes_locatable(ax1)
cax1 = divider1.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im1, cax=cax1, orientation='vertical')

divider2 = make_axes_locatable(ax2)
cax2 = divider2.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im2, cax=cax2, orientation='vertical')

ax1.set_title('Sharpness')
ax2.set_title('Roundness')

plt.show()
```
