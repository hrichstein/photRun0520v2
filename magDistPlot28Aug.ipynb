```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib import rcParams
rcParams["font.family"] = "serif"
```

```python
nonTreas = np.genfromtxt('nonTreas.txt',names=True)
treas = np.genfromtxt('satDistMag.txt',names=True)
lmc = np.genfromtxt('lmcAssoc.txt',names=True)
mw = np.genfromtxt('longMW.txt',names=True)
```

```python
acLMC_d = 49.97
acLMC_mv = -18.1

acSMC_d = 61.7
acSMC_mv = -16.8
```

```python
# 49.97-30 = 19.97
print(14.03+61.7)
```

```python
fig, ax = plt.subplots(figsize=(13,13))

ax.scatter(nonTreas['M_V'],nonTreas['Dist_helio'],marker='^',s=300,\
          color='silver',label='MW Dwarfs')
ax.scatter(treas['M_V'],treas['Dist_helio'],marker='*',\
           color='darkblue',s=500,label='HST-GO 14734')
ax.scatter(mw['M_V'],mw['Dist_helio'],marker='D',\
           color='mediumseagreen',s=300,label='Long-Term MW')
ax.scatter(lmc['M_V'],lmc['Dist_helio'],marker='d',\
           color='mediumpurple',s=300,label='LMC Associated')
ax.plot(acLMC_mv,acLMC_d,marker='o',ms=30,color='black')
ax.plot(acSMC_mv,acSMC_d,marker='o',ms=30,color='mediumpurple')

ax.text(-16.2,75.7,"SMC",size=20)
ax.text(-17.4,63.8,"LMC",size=20)

ax.set_xlim(1,-19)
ax.legend(fontsize=25,loc='best')
ax.set_xlabel(r'M$_V$',fontsize=40)
ax.set_ylabel('Distance (kpc)',fontsize=40)
ax.tick_params(axis='both', which='major', labelsize=20)

# ax.


# plt.show()
plt.savefig('magDist.png',dpi=600,bbox_inches='tight')
```
