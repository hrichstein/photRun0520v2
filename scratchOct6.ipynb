```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# file = np.genfromtxt('photUtils28Sep/catDir_HYDRA-II/magZPTedAll_F606W_0510.dat',names=True)
file = np.genfromtxt('photUtils28Sep/catDir_TUCANA-II-SE/magZPTedAll_F606W_0510.dat',names=True)
```

```python
print(file['magZPTerr']-file['stdev'])
print(np.median(file['magZPTerr']-file['stdev']))
```

```python
fig, ax = plt.subplots()

ax.scatter(file['mean'],file['stdev'],label='stdev',s=5)
ax.scatter(file['mean'],file['magZPTerr'],label='cut',s=5)
ax.scatter(file['mean'],file['magZPTerrfull'],label='full',s=5)

# ax.set_ylim(-0.02,0.05)
#
ax.legend()

plt.show()
```
